## Project Aim and Objectives

According to [Met Office](https://www.metoffice.gov.uk/weather/climate-change/causes-of-climate-change), the main cause of climate change is burning fossil fuels such as oil, gas, and coal. When burnt, fossil fuels release carbon dioxide (CO2) into the air, causing the planet to heat up. This statement can be supported by [NASA](https://climate.nasa.gov/vital-signs/carbon-dioxide/#:~:text=Carbon%20dioxide%20in%20the%20atmosphere,in%20less%20than%20200%20years.), who claims that carbon dioxide (CO2) in the atmosphere warms the planet, causing climate change. Human activities have raised the atmosphere’s carbon dioxide content by 50% in less than 200 years.

For this project, the aim is to find data relaiting to global emmisions of CO2 and other polluting gasses and to analyse the changes in these emmisions before and after the COVID-19 pandemic. Due to the pandemic, many countries implemented measures to allow workers to perform their jobs safely, mainly by allowing workers to work from home. The idea we are working with is that this change in working would likely have some impact of the amount of CO2 emmisions from countries that implemented this strategy due to the increase in demand for electricity with more people working remotely. Additionally, it may be the case that with less people travelling, the emmisions from transportation may have decreased which would be another potentional point to investigate. The overall idea is that we want to understand whether the pandemic had a significant impact on emmision levels. We believe that investigating these results could motivate to look at some sectors and countries in more details in order to make some changes and ,hopefully, slow down the climate change.

We shall take data relating to global emmisions and break it down by country. By combining this data with other data about the countries, such as energy demand, we can see how and why emmisions may have changed for these countries. We will also break down the specific sources of emmisions as there are several areas from which they may be emmited, from transportant to general industrial pollution. All of this data will be required in order to determine exactly where the emmisions are coming from and whether it is related to an increase in energy demand or some other unseen factor.

As will be shown in the Outcome section the codebook is quite long and it could be problematic to find the description of some specific column name. Therefore, function ***codebook_lookup*** is created. It requires 2 variables:

**cb** - the dataset of the codebook represented as pandas dataframe.

**lst** - list of coloumns that the user want to know description of. (if the list is empty returns the whole codebook)

The function returns the dataframe of codebook containing only column names included in the input list.

The example use of this function can be seen in Objective 1 of Outcome section.

```python
def codebook_lookup(cb,lst):
    if lst == []:
        return cb.style.set_properties(**{'text-align': 'left'})

    cb = cb.loc[cb["column_name"].isin(lst)]
    return cb.style.set_properties(**{'text-align': 'left'})
        
```

### Objective 2 code

The function below takes dataframe **df**, integer **start_year**, integer **end_year** and string **variable** and returns the list of the sum of the **variable** grouped by year range. We know that the values in the dataframe are represented as millions. So, for better visualization output values are converted to billions and rounded to 2 decimals.

```python
def sum_var_to_list(df,start_year,end_year,variable):
    mask = (df["year"] >= start_year) & (df["year"] <=end_year)
    temp_df = df.loc[mask]
    output = temp_df.groupby("year")[variable].sum().tolist()

    output = [ elem / 1000 for elem in output ]
    output = [ round(elem,2) for elem in output ]
    return output
```

```python
def back_to_df(input_list,year_range):
    return pd.DataFrame(list(zip(input_list,year_range)))
```

In the code cell below there is a function to plot a graph of annual co2 emission. Variable **general_co2** is calculated using function above.

```python
def plot_ann_co2():
    general_co2 = sum_var_to_list(df,1991,2021,"co2")
    general_year = range(1991,2022,1)

    fig,axs = plt.subplots(figsize=(16,10))

    axs.plot(general_year, general_co2, marker="o")

    axs.set_xticks(range(1991,2022,3))
    axs.set_xlabel("Year")

    fig.suptitle("Annual global emissions of CO2 (billion tons)",fontsize = "x-large")
    return
```

The code below is used to find top 10 countries by co2 emission in pre COVID-19 year(2019) and COVID-19 year(2020).

```python
top = df.sort_values(by=["co2"],ascending=False)
top_pre_covid = top.loc[(top["year"] == 2019)][:10]
top_covid = top.loc[(top["year"] == 2020)][:10]
```

The function below plots 2 graphs of top 10 countries by co2 emission in 2019 and 2020.

```python
def top_10_co2():
    fig, axs = plt.subplots(1,2,figsize=(20,10),sharex=True)
    fig.suptitle("Visualizing top 10 countries by CO2 emissions before and during COVID-19", fontsize= "xx-large")

    axs[0].barh("country","co2",data=top_pre_covid)
    axs[0].invert_yaxis()
    axs[0].set_title("Top 10 countries by CO2 emissions in 2019")
    axs[0].set_xlabel("Carbon dioxide emissions (million tons)")

    axs[1].barh("country","co2",data=top_covid)
    axs[1].invert_yaxis()
    axs[1].set_title("Top 10 countries by CO2 emissions in 2020")
    axs[1].set_xlabel("Carbon dioxide emissions (million tons)")
    return
```

Code below is also used to find top 10 countries in the same time period but this time the value is co2 emission per capita insted just emission in general. To do that the co2 value of the row is multiplied by 1000000 (as it is originally in millions) and then devided by the population of the country. Then the same process as above is used.

```python
top_per_cap = df
top_per_cap["co2_per_capita"] = (top_per_cap["co2"] * 1000000) / top_per_cap["population"]
top_per_cap = top_per_cap.sort_values(by=["co2_per_capita"],ascending=False)
top_cap_pre_covid = top_per_cap.loc[(top_per_cap["year"] == 2019)][:10]
top_cap_covid = top_per_cap.loc[(top_per_cap["year"] == 2020)][:10]
```

The function below used to visualize the top 10 countries by co2 emissions per capita, calculated in the previous cell.

```python
def top_10_per_cap():    
    fig, axs = plt.subplots(1,2,figsize=(25,10),sharex=True)
    fig.suptitle("Visualizing top 10 countries by CO2 per capita emissions before and during COVID-19", fontsize= "xx-large")

    axs[0].barh("country","co2_per_capita",data=top_cap_pre_covid)
    axs[0].invert_yaxis()
    axs[0].set_title("Top 10 countries by CO2 per capita emissions in 2019")
    axs[0].set_xlabel("Carbon dioxide emissions (tons)")

    axs[1].barh("country","co2_per_capita",data=top_cap_covid)
    axs[1].invert_yaxis()
    axs[1].set_title("Top 10 countries by CO2 per capita emissions in 2020")
    axs[1].set_xlabel("Carbon dioxide emissions per capita (tons)")
    return
```
